```python
import sklearn
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('preprocessed.csv')
data = data.drop(['Date'], 1)
n = data.shape[0]
print(data.shape[0], data.shape[1])
data_close = data.values[:,0]
#print(data_close)
```

```python
scaler = MinMaxScaler()
data_slices = []
y_slices = []
for i in range(n-14):
    current_slice = data_close[i:i+14]
    current_slice = scaler.fit_transform(current_slice.reshape(-1,1))
    data_slices.append([ piece[0] for piece in current_slice[:14] ] + [data.values[(i+14), 1]])
    movement = current_slice[-1] - current_slice[-2]
    y_slices.append(0 if np.sign(movement) ==  -1 else 1)

#print(data_slices[1], y_slices[0])
```

```python
X_train = data_slices[:int(n*0.7)]
y_train = y_slices[:int(n*0.7)]
X_test = data_slices[int(n*0.7):]
y_test = y_slices[int(n*0.7):]
```

```python
model = Sequential()
model.add(Dense(12, input_dim=15, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
model.fit(np.array(X_train), np.array(y_train), epochs=100, batch_size=10)
```

```python
stats = []
for i in range(len(X_test)):
    X = np.array(X_test[i])
    X = [[X]]
    y = y_test[i]
    output = model.predict(X)
    print(output, y)
    sign = 1 if output[0] > 0.6 else 0
    stats.append(1 if  sign == y else 0)

print("acc: " + str(sum(stats) / len(stats)))
```
