```python
import pandas as pd
import csv
from datetime import datetime
```

```python
NIFTY = pd.read_csv('NIFTY.csv')
sentiment = pd.read_csv('output_2.csv')
```

```python
daywise = {}
i = 0
for index, row in sentiment.iterrows():
    date = row['Date']
    if date not in daywise:
        daywise[date] = [0, 0]
    if row['Negative Sentiment'] < row['Positive Sentiment']:
        daywise[date][0] += 1
        continue
    daywise[date][1] += 1

print(daywise)
```

```python
day = {}
for index, row in NIFTY.iterrows():
    date = row['Date']
    date = datetime.strptime(date, '%d-%b-%Y').strftime("%Y-%m-%d")
    if date in daywise:
        day[date] = [row['Close'], daywise[date][0] / sum(daywise[date])]

print(day)
```

```python
with open('preprocesed.csv', 'w+', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Date', 'Close', 'Ratio'])
    for key, value in day.items():
        writer.writerow([key, value[0], value[1]])
```
